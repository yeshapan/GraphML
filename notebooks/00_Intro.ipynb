### **Intro to NetworkX**

```python
#import package
import networkx as nx
```

NetworkX provides different classes to store diff types of graphs, such as directed and undirected graph or multigraphs (both directed and undirected)

Read: [NetworkX graph types](https://networkx.org/documentation/stable/reference/classes/index.html).

```python
# Create undirected graph
G = nx.Graph()
print(G.is_directed())

# Create directed graph
H = nx.DiGraph()
print(H.is_directed())

# Add graph level attribute
G.graph["Name"] = "Bar"
print(G.graph)

'''
graph-level attribute is an attribute assigned to entire graph - not a particular node or edge
here, attribute "name" with value "bar" is assigned to the whole graph
'''
```

We can add nodes and edges with attributes in a NetworkX graph

### **Node**

```python
# Add one node with node level attributes
G.add_node('N0', feature=5, label=0)

# Get attributes of the node 0
node_0_attr = G.nodes['N0']
print("Node 0 has the attributes {}".format(node_0_attr))
```

```python
# Add multiple nodes with attributes
G.add_nodes_from([
  ('N1', {"feature": 1, "label": 1}),
  ('N2', {"feature": 2, "label": 2})
]) #(node, attribute_dict)

# Loop through all the nodes
for node in G.nodes(data=True):
  print(node)
'''Set data=True will return node attributes'''

# Get number of nodes
num_nodes = G.number_of_nodes()
print("G has {} nodes".format(num_nodes))
```

### **Edge**

```python
# Add a weighted edge from N0 to N1
G.add_edge('N0', 'N1', weight= 4)

# Get attributed of the edge (0,1)
edge_01_attr = G.edges[('N0','N1')]
print("Edge (0,1) has the attributes {}".format(edge_01_attr))
```

```python
# Add multiple edges with edge weights
G.add_edges_from([
  ('N0', 'N2', {"weight": 7}),
  ('N2', 'N1', {"weight": 2})
])

# Loop through all edges
for edge in G.edges(data=True):
  print(edge)

# Get number of sdges
num_edges = G.number_of_edges()
print("G has {} edges".format(num_edges))
```

### **Visualization**

```python
import matplotlib.pyplot as plt

nx.draw(G, with_labels=True, node_color='pink')

# Get edge labels (weights)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos=nx.spring_layout(G), edge_labels=edge_labels)

plt.show()
```

### **Node degree and Neighbour**

```python
node_id = 'N1'

# Degree of node 1
print("Node {} has degree {}".format(node_id, G.degree[node_id]))

# Get neighbor of node 1
for neighbor in G.neighbors(node_id):
  print("Node {} has neighbor {}".format(node_id, neighbor))
```
